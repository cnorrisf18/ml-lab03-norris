# Logistic Regression
## With intro to Pandas and Graphing

A logistic problem involves making a *logical* decision about some data set.  

Logistic  Regression, as well as regular Linear Regression, is one of the most common supervised learning techniques in the machine language toolset.  It's a *classification* problem - you're trying to decide whether or not to give a certain answer based on certain data.

You have a bunch of inputs, called *features*, and you want to make a decision about them -

You have a set of input variables, or *features*, that all need to be:
* Somewhat independent of one another
* For features that we think are not monotonic (increasing the value either always increases or decreases the probability of a certain outcome) it may help to add a polynomial (x^2) term for those features. You'll see this in the literature with a phrase like "The independent variables are linearly related to the log odds."
* Features should be normalized on their means and standard deviations to make gradient descent work faster.

You want to get a probability that the answer, based on those features, is "yes" - and this result will be value 1.  "no" will be value 0.

For Logistic Regression, we'll do this by determining our model, which is simply a set of linear coefficients to combine all of our features.  Then we have to figure out what the parameters are by training the model.

The features are combined with a *logistic function* that has certain properties - it penalizes error usually with an exponential cost for being "more wrong."

Steps:
* Collect the data
* Clean the data by:
* _ removing unnecsessary features
* _ Filling in any missing values
* _ Filtering for bad / out of range outliers and correcting or removing them
* _ Mapping the features so that increasing the feature linearly increases or decreases the likelihood of a positive result. Add a square term feature is a simple approach, but not a linear one.  Ng doesn't mention this in Coursera.  
* _ Making numbers out of any text or discrete values, possibly with *one hot* approach
* _ Check the data for cross-correlation - are any of the features too dependent on one another?  If so, remove a feature.
* Normalize the data
* Separate the data into at least two data sets:  training, and test.  
* _ the test set is used to measure the progress of the learning algorithm, to measure overfitting (variance - model is too complex and its plotted surface wobbles a lot to overfit the data) and underfitting (bias - model is too simple).  
* _ If you have parameters to tune, such as the learning rate or regularization parameters, you'll need to also generate a *validation* set to use to tune those parameters before applying the test set.
* Train the model
* Apply the model

# Data Preprocessing

We can't train on the categorical values - the ones that are text.  Everything has to be numeric.  That's because the regression (and pretty much every learning method) works on vectors of *numbers*.  What to do?

A "dummy variable" has only two values, zero and one.  What we want to do is convert the categorical values to a bunch of columns of either 0 or 1. This is also called *one hot encoding*.  For some reason Ng doesn't talk at all about this but it is one of the core ideas.

For example take the original "marital" column which had values:

> `single`, `married`, `divorced`

We take that one marital column and make three columns out of it:

> `is_single`, `is_married`, and `is_divorced`

For a row where marital was "married", 

> `is_single` is 0,  `is_married` is 1, and `is_divorced` is 0.

Only **one** of the resulting columns is 1, for the matching category.  All the others are 0.



In logistic regression models, encoding all of the independent variables as dummy variables allows easy interpretation and calculation of the odds ratios - the resulting accuracy of the probability in the prediction - and increases the stability and significance of the coefficients.

pandas has all kinds of things in it to make this work easy.  The one-hot encoding for all the columns that are categorical can be done on *one line* of code.  It'd take you a load of programming to do that otherwise!

Our categorical columns are...
> `['job', 'education', 'default', 'contact', 'poutcome']`

Pandas will even keep the non-categorical ones for us, and drop the ones that get converted to one-hot encoding.  There's nothing else to do!

````python
keep_onehot = pd.get_dummies(keep, columns =['job', 'education', 'default', 'contact', 'poutcome'])
keep_onehot.columns
````
